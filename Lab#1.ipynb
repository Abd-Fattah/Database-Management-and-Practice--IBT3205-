```python
import numpy as np
```

```python
y = df.loc[:49, "MonthlyCharges"].values
x = [i for i in range(0, 50)]

mean_y = np.mean(y)

# Plot the values
plt.plot(x, y, label="Monthly Charges")
plt.axhline(mean_y, color='r', label=f"Mean: {mean_y:.2f}")

# Adding labels and title
plt.xlabel("Index")
plt.ylabel("Monthly Charges")
plt.title("Monthly Charges and Mean")
plt.legend()

# Show plot
plt.show()
```

```python
y_reduced = np.where(y < mean_y, mean_y, y)

plt.plot(x, y_reduced, label="Monthly Charges")
plt.axhline(mean_y, color='r', label=f"Mean: {mean_y:.2f}")

# Adding labels and title
plt.xlabel("Index")
plt.ylabel("Monthly Charges Reduced")
plt.title("Monthly Charges Reduced and Mean")
plt.legend()
```
